```python
import numpy as np 
import matplotlib.pyplot as plt 
import numpy_financial as npf 
# Nr 1: Obliczenie przyszłej wartości mieszkania - future_value_home
PV = 120000 
annual_increase_rate = 0.05 
years = 5 
future_value_home = PV * (1 + annual_increase_rate) ** years

# Nr 2: Obliczenie wymaganej miesięcznej wpłaty
annual_interest_rate = 0.12 
monthly_interest_rate = annual_interest_rate / 12 
months = years * 12
P = npf.pmt(monthly_interest_rate, months, 0, future_value_home)

# Wyświetlenie wyników 
print(f"Przyszła wartość mieszkania za 5 lat: {future_value_home:.2f} zł") 
print(f"Wymagana miesięczna wpłata: {abs(P):.2f} zł") 

months = np.arange(1, years * 12 + 1) 
home_prices = PV * (1 + annual_increase_rate / 12) ** months 
savings = npf.fv(monthly_interest_rate, months, P, 0) 
plt.figure(figsize=(10, 6)) 
plt.plot(months, home_prices, label="Cena Mieszkania") 
plt.plot(months, savings, label="Wartość Lokaty") 
plt.xlabel("Miesiące") 
plt.ylabel("Złote") 
plt.title("Zmiana ceny mieszkania i wartości lokaty") 
plt.legend() 
plt.grid(True) 
plt.show()
savings
```
